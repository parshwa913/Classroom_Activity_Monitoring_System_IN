```python
# ==============================================
# üé¨ Full Colab Script ‚Äî Extract Frames from Video (Google Drive)
# ==============================================

from google.colab import drive
import cv2
import os
import shutil
from google.colab import files

# Step 1Ô∏è‚É£: Mount Google Drive
print("üìÇ Mounting Google Drive...")
drive.mount('/content/drive')

# Step 2Ô∏è‚É£: Set your video path in Drive
# üëá CHANGE this path to match where your video is stored in your Drive
video_path = "/content/drive/MyDrive/Recordingcams.mp4"

# Output folder (frames will be saved here)
output_folder = "frames_1fps"
os.makedirs(output_folder, exist_ok=True)

# Step 3Ô∏è‚É£: Open the video
cap = cv2.VideoCapture(video_path)
fps = cap.get(cv2.CAP_PROP_FPS)

# Fallback in case FPS isn't detected
if fps == 0:
    fps = 30

frame_id = 0
saved = 0

print(f"üéûÔ∏è Starting frame extraction from: {video_path}")
print(f"üïí Detected FPS: {fps}")

# Step 4Ô∏è‚É£: Extract 1 frame per second
while True:
    ret, frame = cap.read()
    if not ret:
        break

    if frame_id % int(fps) == 0:
        frame_name = os.path.join(output_folder, f"frame_{saved:05d}.jpg")
        cv2.imwrite(frame_name, frame)
        saved += 1

    frame_id += 1

cap.release()
print(f"‚úÖ Done! Extracted {saved} frames into '{output_folder}'")

# Step 5Ô∏è‚É£: Zip the frames folder
zip_name = output_folder + ".zip"
shutil.make_archive(output_folder, 'zip', output_folder)
print(f"üì¶ Created ZIP file: {zip_name}")

# Step 6Ô∏è‚É£: Download the zip file
print("‚¨áÔ∏è Preparing ZIP for download...")
files.download(zip_name)

print("üéâ All done! You can now upload the ZIP to Roboflow for annotation.")
```
